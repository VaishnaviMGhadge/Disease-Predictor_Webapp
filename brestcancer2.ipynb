```python
import pandas as pd
import numpy as np
df=pd.read_csv('wisc_bc_data.csv')
df.head(10)
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df['diagnosis'].replace({'M':0,'B':1},inplace=True)
df
```

```python
df.dtypes
```

```python
X=df.drop(['smoothness_mean','symmetry_mean',
           'texture_se','smoothness_se',
           'compactness_se','concave points_se',
           'symmetry_se','smoothness_worst',
           'fractal_dimension_se','diagnosis','id','fractal_dimension_worst'],axis=1)
X
```

```python
y=df['diagnosis']
y
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
select=SelectKBest(score_func=chi2,k='all')
select
```

```python
fit=select.fit(X,y)
```

```python
dfscores=pd.DataFrame(fit.scores_)
dfcols=pd.DataFrame(X.columns)
```

```python
features=pd.concat([dfscores,dfcols],axis=1)
features
```

```python
features.columns=['specs','value']
```

```python
features
```

```python
import seaborn as sns
df.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.10,random_state=5)
```

```python
from sklearn.svm import SVC
model=SVC(kernel='linear')
model
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
model.score(X_test,y_test)
```
