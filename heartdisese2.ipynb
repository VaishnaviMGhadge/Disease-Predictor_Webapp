```python
import pandas as pd
import numpy as np
df=pd.read_csv('datasets_4123_6408_framingham.csv')
df.head(10)
```

```python
df.isnull().sum()
```

```python

df1=df.interpolate()
df1
```

```python
df1.isnull().sum()
```

```python
X=df1.drop(['heartRate','currentSmoker','TenYearCHD'],axis=1)
X
```

```python
y=df1['TenYearCHD']
y
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
select=SelectKBest(score_func=chi2,k='all')
select
```

```python
fit=select.fit(X,y)
dfscore=pd.DataFrame(fit.scores_)
dfcols=pd.DataFrame(X.columns)
```

```python
features=pd.concat([dfscore,dfcols],axis=1)
```

```python
features.columns=['specs','value']
```

```python
features
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=0)
```

```python
from sklearn.svm import SVC
model=SVC(kernel='linear')
model
```

```python
model.fit(X_train,y_train)
```

```python
print(X_test)
```

```python
print(y_test)
```

```python
y_pred=model.predict(X_test)
y_pred
```

```python


from sklearn.metrics import accuracy_score 
```

```python
print(accuracy_score(y_test,y_pred))
```
